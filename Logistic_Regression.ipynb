```python
import pandas as pd
import numpy as np
```

```python
titanic_data = pd.read_csv('titanic.csv')
```

```python
titanic_data.head()
```

```python
titanic2 = titanic_data[['pclass', 'survived', 'sex', 'age']]
titanic2.head()
```

```python
titanic2.isnull().sum()
```

```python
titanic2.shape
```

```python
titanic3 = titanic2.dropna()
```

```python
titanic3
```

```python
titanic3.isnull().sum()
```

```python
titanic4 = pd.get_dummies(titanic3, columns=['sex'])
```

```python
titanic4
```

```python
y = titanic4[['survived']]
X = titanic4[['pclass', 'age', 'sex_female']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.85, random_state = 7)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
# building model in trainning data 
from sklearn.linear_model import LogisticRegression 
LR = LogisticRegression()
model_lr = LR.fit(X_train, y_train)
model_lr
```

```python
y_test['Predict'] = model_lr.predict(X_test)
```

```python
y_test
```

```python
# confusion matrix
pd.crosstab(index = y_test['survived'], columns = y_test['Predict'], margins = True)
```
