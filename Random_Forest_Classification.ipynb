```python
import pandas as pd
mobile_data = pd.read_csv("Mobile Classification.csv")
```

```python
avg_ram = mobile_data.groupby('price_range')[['ram']].mean().reset_index()
```

```python
avg_ram
```

```python
y = mobile_data[['price_range']]
X = mobile_data.drop(columns = ['price_range'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .25, random_state = 42)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 1000, random_state = 42)
rf = model.fit(X_train, y_train)
```

```python
rf
```

```python
y_test['Prediction'] = model.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test['price_range'], y_test['Prediction']))
print(accuracy_score(y_test['price_range'], y_test['Prediction']))
```

```python
forest_importance = pd.Series(model.feature_importances_, index = X.columns)
```
